```python
from MontyHallSimulation import monty_hall
import matplotlib.pyplot as plt

%config InLineBackend.figure_format = 'retina' #instead of png
```

```python
# special function to get list of winning percentage after each game
def simulation(switch, n):
    wins = 0
    wp_list = []

    for i in range(n):
        if monty_hall(False, switch):  # not interactive and switch based on arg
            wins += 1
        wp_list.append(wins/(i+1))

    winning_percent = round((wins/n)*100, 2)
    return wp_list
```

```python
stay = simulation(switch=False, n=1000)
```

```python
switch = simulation(switch=True, n=1000)
```

```python
games = []
for i in range(1000):
    games.append(i)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6))

ax.plot(games, switch, lw = 3, color = 'red') # same as ax.plot(avg_mpg)
ax.plot(games, stay, lw = 3) # same as ax.plot(avg_mpg)

ax.set_xlabel("Games Played")
ax.set_ylabel("Win Percentage")
ax.set_title("Monty Hall Simulation", size = 20)
ax.text(880, 0.7,"Switched", color = 'red', size = 15)
ax.text(905, 0.37,"Stayed", color = '#1f77b4', size = 15)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(.5)
ax.spines['bottom'].set_linewidth(.5)
plt.show()
```
